# 데이터 전처리

정리해야 할 것들
1. 데이터 중에서 노란색 셀 부분만 사용한다.<br>
2. 특징이 총 9개(0~8)<br>
3. 각 특징에 데이터 차례대로 넣기<br>
4. 이때 아무것도 없는 값은 지우기 ( Nan )<br>

클래스 -> 큰특징(0~9) -> 세부특징(0~9)<br>

```python
import openpyxl
import pandas as pd
import numpy as np

# feature_0 = [ [ [0-0] ~ [0-8] ], ~ , [ [8-0] ~ [8-8] ] -> [9][9][append]
xlsx = openpyxl.load_workbook('mCTask.xlsx')
FinalData = pd.DataFrame()

for cls in ['0', '1', '2' ,'3']:
    clss = xlsx.get_sheet_by_name(cls)
    
    usful = [] # 노랑색 셀의 행번호를 저장할 List

    for i in range(8, 100):
        k = 'B'+ str(i) # B8
        cell = clss[k] # clss['B8']
        if(cell.fill.start_color.index == 'FFFFFF00'): # 'FFFFFF00' = Yellow
            usful.append(i-1)

    data = pd.DataFrame(clss.values)
    data = data.iloc[usful,4:]
    data = data.dropna(how = 'all', axis = 1) 
    data = data.dropna(how = 'all', axis = 0)
    data = data.reset_index(drop=True)
    # 결측치 처리 # Nan -> Mean #
    data = SimpleImputer(missing_values=np.nan, strategy='mean').fit_transform(data) # mean 값으로 결측치 해결하기
    data = pd.DataFrame(data)
    data['target'] = cls
    FinalData = pd.concat([FinalData, data],ignore_index=True)

FinalData.columns = [i for i in range(1,83)]
FinalData.rename(columns={82:"target"}, inplace = True) 

X = FinalData.iloc[:,:81]
display(X)
Y = FinalData['target']
```

# 2. 데이터 분류하기 ( Train / Test )

OverFitting = 차이가 0.2이상<br>
UnderFitting = 점수가 0.4이하<br>

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import LeaveOneOut 

from imblearn.over_sampling import * # 과대표준화 하기위한 라이브러리 (Over Sampler)
from sklearn.svm import SVC 

#Leave-One-Out 교차 검증을 위한 함수 Load #
loo = LeaveOneOut()
loo.get_n_splits(X)

best_score_te = 0
best_score_tr = 0
loo_s= 0 # 최대값
loo_c = 0 # 최대 C값
loo_g = 0 # 최대 gamma값
loo_t_train = 0 # train 토탈 값
loo_t_test = 0 # test 토탈 값

for c in [0.001, 0.01, 0.1, 1, 10]:
    for gamma in [0.001, 0.01, 0.1, 1, 10]:

        #================== Leave One Out =======================================================#
        for train_index_loo, test_index_loo in loo.split(X):

            # 2. 학습데이터 / 시험데이터 분류하기 #
            X_train, X_test = X.iloc[train_index_loo], X.iloc[test_index_loo] # 학습데이터 , 테스트데이터
            Y_train, Y_test = Y[train_index_loo], Y[test_index_loo] # 학습 타겟 , 테스트 타겟

            # 과대표준화(Over-Sampleing) 작업 #
            # 이때 생성된 균형샘플은 평가에 영향을 미치면 안된다. #
            X_train_os, Y_train_os = RandomOverSampler().fit_sample(X_train, Y_train) 

            # 3.분류기 학습 #
            # 과대표준화(OverSampled) 된 (학습데이터와 학습타겟)을 학습시킨다.
            svm = SVC(kernel='rbf', C=c, gamma = gamma).fit(X_train_os, Y_train_os)

            loo_t_test += svm.score(X_test, Y_test) # Leave-One-Out 방식은 총 성공, 실패를 더한 값을
            loo_t_train += svm.score(X_train, Y_train)
        score_loo_te = loo_t_test / len(X) # 전체 데이터 X의 개수(160개)로 나눈다.
        score_loo_tr = loo_t_train / len(X)

        # 최적의 파라미터 값을 찾아내기 위한 조건 연산 #
        if(best_score_te < score_loo_te):
            best_score_te = score_loo_te
            best_score_tr = score_loo_tr
            loo_c = c
            loo_g = gamma
        loo_t_test = 0
        loo_t_train = 0

print("===== Leave-One-Out =====")
print("최대값(test)   : {0}".format(best_score_te))
print("최대값(train)   : {0}".format(best_score_te))
gap = best_score_tr - best_score_te
print("차이(gap) : {0}".format(gap))

if(gap >= 0.2):
    print("상태 : Over-Fitting")
elif(best_score_te <= 0.4):
    print("상태 : Under-Fitting")
else:
    print("상태 : Balanced")
    

print("C값      : {0}".format(loo_c))
print("Gamma값  : {0}".format(loo_g))
print()
```
